```python
# Intern 6: Function for water needs (0.5L per °C above 20°C).

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("strawberrywithmois.csv")
print("\n------------------ GIVEN DATA SET ------------------")
print(df.to_string(index = False))

def water_needs(temperature):
    if temperature > 20:
        return (temperature - 20) * 0.5
    else:
        return 0

def main():
  df['water_needs'] = df['temperature'].apply(water_needs)
  new_df = df[['plant_id', 'temperature', 'water_needs']]
  print("\n------------------ RESULT DATA ------------------")
  print(f"{new_df.to_string(index=False)}")

  plt.figure(figsize=(8,6))
  plt.bar(new_df['plant_id'], new_df['water_needs'], color='orange')
  plt.xlabel('Plant ID')
  plt.ylabel('Water Needs (liters)')
  plt.title('Water Needs for Each Plant')
  plt.grid(axis = 'y', linestyle ='dashed', alpha =0.7)
  plt.tight_layout()
  plt.savefig("water_needs_plot.png")
  plt.show()


  while True:
    output = input("\nDownload o/p data: Type 1 for csv, 2 for Excel, 3 to Decline --- ")

    if output == "1":
        new_df.to_csv("water_needs.csv", index=False)
        break
    elif output == "2":
        new_df.to_excel("water_needs.xlsx", index=False)
        break
    elif output == "3":
        break
    else:
        print("!! Invalid Output !!")


if __name__ == "__main__":
    main()
```
